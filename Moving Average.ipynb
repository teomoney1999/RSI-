```python
import pandas as pd
import matplotlib as plt
%matplotlib inline
```

```python
btc = pd.read_csv('NASDAQOMX-XQC.csv')
```

```python
btc = btc.set_index('Trade Date')
```

```python
btc = btc.sort_values(by='Trade Date')
```

```python
price_line = btc['Index Value'].plot(figsize=(10,10))
```

```python
btc['20-day MA'] = btc['Index Value'].rolling(20).mean()
```

```python
btc[['Index Value', '20-day MA']].plot(figsize=(10,10))
```

```python
btc['5-day MA'] = btc['Index Value'].rolling(5).mean()
```

```python
btc[['Index Value', '5-day MA', '20-day MA']].head(100).plot(figsize=(10,10))
```

```python
btc[['Index Value', '5-day EMA', '20-day EMA']].head(200).plot(figsize=(10,10))
```

```python
btc['20-day EMA'] = btc['Index Value'].ewm(span=20).mean()
```

```python
btc[['Index Value', '20-day EMA']].head(100).plot(figsize=(10,10))
```
